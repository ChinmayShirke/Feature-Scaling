```python
miss_cat_var=[var for var in catvar.columns if catvar[var].isnull().sum()>0]
miss_cat_var
```

```python
data=df.copy()
for var in miss_cat_var:
    data.update(df[var].fillna( df[var].mode()[0]))
    
```

```python
data.isnull().sum()
```

```python
numvar=data.select_dtypes(include=['float','int'])
miss_num_var=[var for var in numvar if numvar[var].isnull().sum()>0]
miss_num_var
```

```python
numvars=['Price','Discount']
catvars=['Course ','Location']
for numvar,catvar in zip (numvars,catvars):
    for unival in data[catvar].unique():
        mask=(data.loc[:,catvar]==unival)
        mean=data.loc[mask,numvar].mean()
        data.update(data.loc[mask,numvar].replace(np.nan,mean))
```

# Spliting Data

```python
from sklearn.preprocessing  import StandardScaler
from sklearn.preprocessing  import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
X=data.drop('Price',axis=1)
Y=data['Price']
print(f"X shape:{X.shape}")
print(f"Y shape:{Y.shape}")
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=51)
print(f'Shape of X_train :{X_train.shape}')
print(f'Shape of X_test :{X_test.shape}')
print(f'Shape of Y_train :{Y_train.shape}')
print(f'Shape of Y_test :{Y_test.shape}')
```

# Standardization

```python
sc=StandardScaler()
sc.fit(X_train)

```

```python
X_train.describe()
```

```python
X_train_sc=sc.transform(X_train)
X_test_sc=sc.transform(X_test)
X_test_sc
```

```python
X_train_sc=pd.DataFrame(X_train_sc,columns=['Discount', 'location', 'course'])
X_test_sc=pd.DataFrame(X_test_sc,columns=['Discount', 'location', 'course'])
```

```python
X_train_sc.describe().round(2)
```

# Normalization

```python
#fitting
mmc=MinMaxScaler()
mmc.fit(X_train)
```

```python
#transforming
X_train_mmc=mmc.transform(X_train)
X_test_mmc=mmc.transform(X_test)
```
